<h2>Hedging Volatility ETFs</h2>
<p>
Low risk hedging strategies can extract the roll return from funds trading futures on the same underlying. The largest yields I've found are in volatility funds, but this also works for commoditites, stock index funds, and anywhere else returns convexity exists.
<p>
Trading leveraged funds on the short side can increase yields, and hedging with inverse volatility funds captures the most convexity.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from zipline.utils.factory import load_bars_from_yahoo
```

```python
end = pd.Timestamp.utcnow()
start = end - 2500 * pd.tseries.offsets.BDay()

data = load_bars_from_yahoo(stocks=[
    'SVXY', 'XIV', 'VXX', 'UVXY', 'TVIX','ZIV', # Volatility
    'SPXL', 'SPY',                              # S&P 
    'TMF', 'TMV',                               # Leveraged Bonds
    'UGAZ', 'UNG',                              # Commodities (nat gas)
    ],
    start=start, end=end)

prices = data.minor_xs('price').dropna()
returns = np.log(prices / prices.shift(1))
```

<h4>The strategy is simple, short volatility and hedge by shorting inverse volatility</h4>
<p>
Here's a demo of the concept using VXX & XIV. The idea is to take two highly anticorrelated return series and compose them together such that the resulting distribution is highly skewed towards the positive side.
<p>
VXX buys front month VIX contracts and sells forward contracts, and XIV buys foward contracts and sells front month contracts. Due to contango in the volatility market, VXX has a large negative roll return, and XIV earns a small positive roll return. The roll returns earned by these funds can be hedged out with very little risk by shorting them both.


```python
# UNROLLING RETURNS
def compound(R):
    return (1 + R).cumprod()

fig, axes = plt.subplots(3, 1, figsize=(12,10))

axes[0].plot(returns[['XIV','VXX']])
axes[0].set_title("XIV/VXX Returns (high negative correlation)")

hedged_returns = -(returns.VXX + returns.XIV)
axes[1].plot(hedged_returns)
axes[1].set_title("Returns Shorting Both")

axes[2].plot(compound(hedged_returns))
axes[2].set_title("Cumulative Returns")

fig.tight_layout()
```

<h4>Convexity yields of different funds and asset classes</h4>
This idea can be used all over, but is the most pronounced in the volatility market. All of these strategies would be profitable, but some would need to take on a lot of leverage. Volatility looks profitable even at Reg-T margin levels.

```python
spreads = pd.DataFrame({
    'VXX/XIV': -(returns.VXX + returns.XIV),
    'TMF/TMV': -(returns.TMF + returns.TMV),
    'SPXL/SPY':-(returns.SPXL/3 - returns.SPY),
    'UVXY/VXX':-(returns.UVXY/2 - returns.VXX),
    'TVIX/VXX':-(returns.TVIX/2 - returns.VXX),
    'UVXY/XIV':-(returns.UVXY/2 + returns.XIV),
    'TVIX/XIV':-(returns.TVIX/2 + returns.XIV),
    'UGAZ/UNG':-(returns.UGAZ/3 - returns.UNG),
    'UVXY/SVXY': -(returns.UVXY/2 + returns.SVXY),
})
yields = spreads.apply(compound)
yields.plot(figsize=(12,8))
```

Hedging short positions on leveraged volatility with short positions on inverse volatility is by far the best performer.
<h3>Selelcting a Risk/Return Profile</h3>
<p>
This is the UVXY/SVXY trade with -50% on the UVXY leg and several different weightings for XIV. It's really simple to fine tune the risk/return profile.

```python
def make_portfolios(y, x, wy, weights):
    return pd.DataFrame({weight: wy*y + weight * x
                         for weight in weights})

weights = np.linspace(-1.5, -0.25, 25)

portfolios = make_portfolios(returns.UVXY, returns.SVXY, -0.5, weights).dropna()
cumR = compound(portfolios)
cumR.plot(figsize=(12,8), legend=False)
```

```python
def rolling_kelly(df, window, rfr=0.0):
    mu = pd.rolling_mean(df, window) - rfr
    var = pd.rolling_var(df,window)
    return mu / var

kellys = rolling_kelly(portfolios, 60)
winners = kellys.mean().nlargest(10)

fig = kellys[winners.index].plot(figsize=(12, 8), legend=1)
fig.set_title('60 Day Rolling Kelly Leverages')
print "10 Best average 60 day Kelly criterion leverage\n" 
print winners, '\n'
```

```python
kelly = portfolios.mean() / portfolios.var()
kelly[kelly > 10].plot()
```

<h4>Backtesting</h4>
<p>
This is the result of re-hedging a portfolio at -50% UVXY and -95% SVXY. I over-estimated commissions to account for 

```python
from zipline.api import (add_history, 
                         history, 
                         set_slippage, 
                         slippage,
                         set_commission, 
                         commission, 
                         order_target_percent)

from zipline import TradingAlgorithm


def initialize(context):
    set_slippage(slippage.FixedSlippage(0))
    set_commission(commission.PerShare(cost=0.02, min_trade_cost=1.5))
    
def handle_data(context, data):
    order_target_percent('UVXY', -0.50)
    order_target_percent('SVXY', -0.95)
        
# Instantinate algorithm        
algo = TradingAlgorithm(initialize=initialize, 
                        handle_data=handle_data)
# Run algorithm
results = algo.run(prices)
results.portfolio_value.plot()
```

```python
returns_regression = pd.ols(y=range(len(results.portfolio_value)), x=results.portfolio_value)
returns_regression
```
