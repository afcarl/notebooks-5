We'll be working with SPY and the SPDR sector ETFs to keep the number of moving parts to a minimum. First, we need to download historical data, Quantopian's zipline library make this a simple task.

```python
%pylab inline
import datetime
import pandas as pd
from zipline.utils.factory import load_bars_from_yahoo

# Get 4000 business day window of prices, we can downsample this later.
lookback = 4000 * pd.tseries.offsets.BDay()

end = pd.Timestamp.utcnow()
start = end - lookback

# Pandas Panel of OHLCV data
data = load_bars_from_yahoo(
    stocks=['XLP', 'XLE', 'XLK', 'XLY', 'XLU', 'XLB', 'XLF', 'XLV','XLI', 'SPY'],
    start=start, end=end
)
prices = data.minor_xs('price')
print data
```

<p>
Recently, the energy sector has been dragged down with oil prices, and underperformed relative to the SPY. Is this level of separation common? Is it likely to revert? In order to investigate further, plotting the historical 100 day spreads between SPY and each sector will put the current spread into a historical context.

<p>
For this to make sense, it is assumed that the sector ETFs prices are linearly related to the price of of the SPY. It's a big assumption, but model simplicity can be under appreciated, linear models can explain a large portion of data variance with little added complexity.
<p>
The spreads are calculated using a 100 day rolling regression as follows. For even more simplicity, no constant term is used.
</p>
<p>
\begin{eqnarray*} SPY &=& h*SECTOR \\
                    0 &=& SPY - h*SECTOR \\
               spread &=& SPY - h*SECTOR
\end{eqnarray*}
<p>
where $h$ is called the hedge ratio for that sector.

```python
# Building the regression models...
WINDOW = 100

X = prices[[i for i in prices if i != 'SPY']]
y = prices.SPY

models = {
    stock: pd.ols(y=y, x=X[stock], 
                  window=WINDOW, 
                  window_type='rolling', 
                  intercept=False)
    for stock in X.columns
}
fitted_X = pd.DataFrame({stock: models[stock].beta['x'] * X[stock] 
                         for stock in models})

spreads = y - fitted_X

spreads.plot(figsize=(12,8))
```

<p>The spreads tend to stay in the $\pm 10$ range, with larger deviations occurring around times of higher volatility. In the upper right of the plot, it looks like XLE is currently well outside of its normal range. 
<p>

<h3>The spread between SPY and XLE is near an all time high!</h3>
<p>
We have a particular trade in mind now, buy XLE and sell SPY. The spread looks to be mean reverting, and it's at or near an all time high. Next the concept of an over/under valued spread needs to be refined so it can be tested with historical data. 
<p>
Using support/resistance bars in an automated strategy is somewhat impractical due to differences across sectors, and drift over time. A simple solution is to assume the spreads fall on a normal Gaussian distribution, and to use the z-score. 
<p>
A spread will be defined as over-valued when its z-score greater than 2, and under-valued when it is below -2. This general definition allows us to move onto the next phase, backtesting.

```python
def rolling_zscore(df, window):
    return (df - pd.rolling_mean(df, window)) / pd.rolling_std(df, window)

zscores = rolling_zscore(spreads, WINDOW)
zscores.XLE.plot(figsize=(13,9))
```

<h3>Backtesting</h3>
<p>
For backtesting purposes we will be using zipline, Quantopian's open source backtesting engine. Zipline is a simulation style backtesting engine, it dynamically steps itself through time, simulating each transaction as it goes. This style of backtesting makes it more difficult to introduce look-ahead bias, and much easier to roll into a live trading environment via Quantopian. 
<p>
The strategy will be tested over the entire the 4000 day window.


```python
import zipline
from zipline.api import *
from zipline import TradingAlgorithm
```

```python
def initialize(context):
    '''
    Called once at the very beginning of a backtest (and live trading). 
    Use this method to set up any bookkeeping variables.
    
    The context object is passed to all the other methods in your algorithm.

    Parameters

    context: An initialized and empty Python dictionary that has been 
             augmented so that properties can be accessed using dot 
             notation as well as the traditional bracket notation.
    
    Returns None
    '''
    add_history(1000, '1d', 'price')
    # Turn off the slippage model
    set_slippage(slippage.FixedSlippage(spread=0.0))
    # Overestimate commissions to account for slippage & unknown costs
    set_commission(commission.PerShare(cost=0.015, min_trade_cost=1.2))
    context.stocks = [
       'XLP', 'XLE', 'XLK', 'XLY', 'XLU', 'XLB', 'XLF', 'XLV','XLI'
    ]
    context.index = symbol('SPY')
    context.max_z = 2
    context.min_z = -2
    context.bet_shares = 100 # Shares of the spread to buy
    context.window = 100
    context.tick = 0

def handle_data(context, data):
    '''
    Called when a market event occurs for any of the algorithm's 
    securities. 

    Parameters

    data: A dictionary keyed by security id containing the current 
          state of the securities in the algo's universe.

    context: The same context object from the initialize function.
             Stores the up to date portfolio as well as any state 
             variables defined.

    Returns None
    '''
    # Allow the data panel to have 100 days of prices before trading.
    context.tick += 1
    if context.tick < 100:
        return
    prices = history(1000, '1d', 'price')
    y = prices[context.index]
    models = {
        stock: pd.ols(y=y, x=prices[stock], 
                  window=context.window, 
                  window_type='rolling', 
                  intercept=False)
        for stock in context.stocks
    }
    fitted_x = pd.DataFrame({stock: models[stock].beta['x'] * prices[stock] 
                             for stock in context.stocks})
    spreads = y - fitted_x
    z = zscore(spreads).iloc[-1]
    
    # Store the orders in a dict, this is mostly to 
    # cancel out any SPY orders with opposite signs.
    bets = {stock: 0 for stock in data}
    for stock in data:
        if stock == context.index:
            continue
        if z[stock] > context.max_z:
            bets[stock] = fitted_x[stock][-1]
            bets[context.index] -= data[context.index].price
        elif z[stock] < context.min_z:
            bets[stock] = -fitted_x[stock][-1]
            bets[context.index] += data[context.index].price
    for stock in data:
        order_target_value(stock, bets[stock] * context.bet_shares)
    # record the range of zscores
    record(z=z.mean(),
           zmax=z.max(),
           zmin=z.min())
    
zscore = lambda x: (x - x.mean()) / x.std()


algo = TradingAlgorithm(initialize=initialize, 
                        handle_data=handle_data)

results = algo.run(prices)
```

```python
results.portfolio_value.plot(figsize=(12,8))           
```

<p>
Analyzing the cash levels of this strategy indicate that it remains very close to market neutral. It borrows the shares of one security, sells them, and buys the other security with the raised cash.
