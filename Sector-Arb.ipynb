<H1>Searching for Opportunity in Market Corrections</H1>
<p>
The market is in the midst of a correction, investors are nervous, and volatility is on the rise... Now what?
<p>
Corrections are a necessity, and markets wouldn't be very exciting without them. Instead of fighting the inevitable, why not develop a strategy around it. To me, volatility is synonymous with confusion and uncertainty. Selloffs are a shock to the market systems, and prices behave erratically while the system seeks its new position of homeostasis. How wild it can go is another story, but experience (and some optimism) suggest that things will settle down again.

<P>
This is an attempt at developing a trading strategy, my goal is to demonstrate the idea development process, not to argue the validity of any strategy. I'll be looking for inter-sector arbitrage opportunities in the S&P 500 during times of higher volatility (confusion).
<p>First, I'll need some data.

```python
%pylab inline
import datetime
import pandas as pd
from zipline.utils.factory import load_bars_from_yahoo
```

```python
# Get 4000 business day window of prices, we can downsample this later.
lookback = 4000 * pd.tseries.offsets.BDay()

end = pd.Timestamp.utcnow()
start = end - lookback

# Pandas Panel of OHLCV data
data = load_bars_from_yahoo(
    stocks=['XLP', 'XLE', 'XLK', 'XLY', 'XLU', 'XLB', 'XLF', 'XLV','XLI', 'SPY'],
    start=start, end=end
)
prices = data.minor_xs('price')
```

<p>
In order to visualize the sector prices relative to the SPY, I normalize the prices by dividing out the price on the start date, and subtract out the SPY component. 

```python
def normed_prices(prices, window):
    return prices.iloc[-window:] / prices.iloc[-window]

# Normalize the last 100 days prices
norm_prices = normed_prices(prices, 100)

# Subtract SPY out to see the relative performance
norm_prices = norm_prices - norm_prices.SPY

norm_prices.plot(figsize=(12,8))
```

<h3> It looks like XLE has deviated significantly from the rest of the pack over the last 100 days.</H3>
<p> I'd like to look at the spreads of the sectors vs. SPY via a rolling 100 day regression. I'll assume that the SPY can be replicated by any of the sectors if we multiply by a constant. The spread will be calculated as follows

\begin{eqnarray*} SPY &=& h*SECTOR \\
                    0 &=& SPY - h*SECTOR \\
                  spread &=& SPY - h*SECTOR
\end{eqnarray*}
<p>
where $h$ is the hedge ratio for that sector.

```python
# Build  Regression models
X = prices[[i for i in prices if i != 'SPY']]
y = prices.SPY

# Rolling window length
window = 100
models = {
    stock: pd.ols(y=y, x=X[stock], 
                  window=window, 
                  window_type='rolling', 
                  intercept=False)
    for stock in X.columns
}

fitted_x = pd.DataFrame({stock: models[stock].beta['x'] * X[stock] 
                         for stock in models})

spreads = y - fitted_x

spreads.plot(figsize=(15,9.5))
```

<p>It looks like the spreads rarely deviate outside of +-20, and on the right, it looks like XLE is out of whack currently. 

```python
spreads.XLE.plot(figsize=(13,9))
```

<h3>The spread between SPY and XLE is at an all time high!

```python
# Get the Z-score of spreads
def rolling_zscore(df, window):
    return (df - pd.rolling_mean(df, window)) / pd.rolling_std(df, window)

zscores = rolling_zscore(spreads.XLE, window)

zscores.plot(figsize=(13,9))
```

<p>The zscore of the spread is in the statistically significant range, but what does that mean? It would be interesting to look at what would have happened in the past if we had bought or sold the spread any time it went outside of 2 standard deviations from its mean.
<p>
I'll use Quantopian's open source backtesting engine, zipline, to backtest the strategy over the last decade. I only expect it to trade during times of higer volatility when market confusion is at a peak.

<h3>Backtesting</h3>

```python
import zipline
from zipline.api import *
from zipline import TradingAlgorithm
```

```python

def initialize(context):
    add_history(1000, '1d', 'price')
#     set_slippage(slippage.FixedSlippage(spread=0.05))
    set_commission(commission.PerShare(cost=0.01, min_trade_cost=1.0))
    context.stocks = [
       'XLP', 'XLE', 'XLK', 'XLY', 'XLU', 'XLB', 'XLF', 'XLV','XLI'
    ]
    context.index = symbol('SPY')
    context.max_z = 2
    context.min_z = -2
    context.bet_shares = 100 # Shares of the spread to buy
    context.window = 100
    context.tick = 0

def handle_data(context, data):
    context.tick += 1
    if context.tick < 100:
        return
    prices = history(1000, '1d', 'price')#.dropna(axis=1)
    y = prices[context.index]
    models = {
        stock: pd.ols(y=y, x=prices[stock], 
                  window=context.window, 
                  window_type='rolling', 
                  intercept=False)
        for stock in context.stocks
    }
    fitted_x = pd.DataFrame({stock: models[stock].beta['x'] * prices[stock] 
                             for stock in context.stocks})
    spreads = y - fitted_x
    z = zscore(spreads).iloc[-1]
    bets = {stock: 0 for stock in data}
    for stock in data:
        if stock == context.index:
            continue
        if z[stock] > context.max_z:
            bets[stock] = fitted_x[stock][-1]
            bets[context.index] -= data[context.index].price
        elif z[stock] < context.min_z:
            bets[stock] = -fitted_x[stock][-1]
            bets[context.index] += data[context.index].price
    for stock in data:
        order_target_value(stock, bets[stock] * context.bet_shares)
    # record the range of zscores
    record(z=z.mean(),
           zmax=z.max(),
           zmin=z.min())
    
zscore = lambda x: (x - x.mean()) / x.std()


algo = TradingAlgorithm(initialize=initialize, 
                        handle_data=handle_data)

results = algo.run(prices)

```

```python
results.portfolio_value.plot(figsize=(12,8))           
```

```python
results[['z', 'zmax', 'zmin']].plot(figsize=(12,8))
```
