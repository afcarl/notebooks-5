```python
%pylab inline
import numpy as np
import cvxopt as opt
from cvxopt import blas, solvers
from pylab import plot ,show, xlabel, ylabel

# Turn off progress printing 
solvers.options['show_progress'] = False

## NUMBER OF ASSETS
n = 4

## NUMBER OF OBSERVATIONS
nobs = 1000

def rand_weights(n):
    ''' Produces n random weights that sum to 1 '''
    k = np.random.randn(n)
    return k / sum(k)

def make_returns(asset_count, nobs, drift=0.0):
    ''' 
    Creates normally distributed series 
    :params:
        asset_count: <int> number of series to create
        nobs: <int> number of observations
        drift: <float> skews the distribution to one side
    :returns:
        np.ndarray with <asset_count> rows and <nobs> columns
    '''
    return np.random.randn(asset_count, nobs) + drift
    
## CREATING PORTFOLIOS FUNCTION
def random_portfolio(returns, weight_func):
    ''' 
    Returns the mean and standard deviation of returns for a random portfolio
    '''
    w = weight_func(returns.shape[0])
    mu = np.dot(np.mean(returns, axis=1) , w)
    sigma = math.sqrt(np.dot(w, np.dot(np.cov(returns), w)))
    # This recursion reduces outliars to keep plots pretty
    if sigma > 2:
        return random_portfolio(returns, weight_func)
    return sigma, mu


## RUNNING AND PLOTTING PORTFOLIOS

return_vec = make_returns(n, nobs, drift=0.0)
stds, means = np.column_stack(
    [random_portfolio(return_vec, rand_weights) for _ in xrange(500)]
)

plot(stds ,means , 'o')
ylabel('mean')
xlabel('std')

## CONVERSION TO MATRIX
k = np.array(return_vec)
S = opt.matrix (np.cov(k))
pbar = opt.matrix(np.mean(k, axis=1))

# ## CONDITIONING OPTIMIZER
G = -opt.matrix(np.eye(n))   # negative n x n identity matrix
h = opt.matrix(0.0, (n ,1))
A = opt.matrix(1.0, (1, n))
b = opt.matrix(1.0)

# ## CALCULATING EXPECTED RETURNS
N = 100
mus = [10**(5.0* t/N - 1.0) for t in range(N)]

# ## CALCULATING THE WEIGHTS FOR EFFICIENT FRONTIER
portfolios = [
    solvers.qp(mu*S, -pbar, G, h, A, b)['x'] 
    for mu in mus
]

## CALCULATING RISKS AND RETURNS FOR FRONTIER
returns = [blas.dot(pbar, x) for x in portfolios]
risks = [np.sqrt(blas.dot(x, S*x)) for x in portfolios]
plot(risks, returns, 'y-o')

## PLOTTING THE KELLY PORTFOLIOS
def get_kelly_portfolios():
    ww = np.dot(np.linalg.inv(opt.matrix(S)), opt.matrix(pbar))
    rks = []; res = [];
    for i in arange (0.05, 20, 0.0001):
        w = ww / i
        rks.append(blas.dot(pbar, opt.matrix(w)))
        res.append(np.sqrt(blas.dot(opt.matrix(w), S*opt.matrix(w))))
    return res, rks

res, rks = get_kelly_portfolios()
    
## PLOTTING THE KELLY PORTFOLIOS FOR DIFF LEVERAGE
plot(res, rks, 'ko', markersize=3)
plot(res, np.array(rks) * -1, 'ko', markersize=3)
show ()

## FITTING THE MARKOWITZ PARABOLA
m1 = np.polyfit(returns, risks, 2)
x = np.linspace(min(returns), max(returns)*2 ,1000)
y = np.polyval(m1, x)

## FITTING THE KELLY WEIGHTS
m2 = polyfit(returns, risks, 1)

# m1 IS THE PARABOLIC FIT AND m2 THE LINEAR FIT
# print m1, m2, ww

## PLOTTING THE TANGENT TO THE PARABOLA
plot(x, (2*(sqrt(m1[0]*m1[2])) + m1[1])*x, 'k')

# ## PLOTTING THE KELLY PORTFOLIOS FOR DIFFERENT FACTORS
plot(rks, res, 'x')

# ## PLOTTING THE MARKOWITZ PARABOLA
plot(returns, risks, 'r-o')

## PLOTTING THE FITTING FUNCTION FOR THE MARKOWITZ PARABOLA
plot(x, y)

## FINDING OPTIMAL MARKOWITZ POINTS
x1 = sqrt(m1[2] / m1[0])
x2 = -sqrt(m1[2] / m1[0])
y1 = m1[0] * x1**2 + m1[1] * x1 + m1[2]
y2 = m1[0] * x2**2 + m1[1] * x2 + m1[2]

## PLOTTING THE OPTIMUM MARKOWITZ POINT
plot([x1], [y1], 'ko', markersize=10)
xlabel('mean')
ylabel('std')

## FINDING OPTIMUM PORTFOLIO WITH CVXOPT
wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']

plot(
    blas.dot(pbar, opt.matrix(wt)), 
    np.sqrt(blas.dot(opt.matrix(wt), S*opt.matrix(wt))), 
    'go', markersize=10
)
show()
```

```python
%pylab inline
import numpy as np
import cvxopt as opt
from cvxopt import blas, solvers
from pylab import plot ,show, xlabel, ylabel
import pandas as pd
from zipline.sources.data_frame_source import DataFrameSource
from zipline.utils import tradingcalendar


class MonteCarloSource(DataFrameSource):

    def __init__(self,
                 vol=.5,
                 rate=.0535,
                 price_range=(5,300),
                 trials=20,
                 time=1000,
                 start=pd.to_datetime('1990-01-01'),
                 **kwargs):

        dates = tradingcalendar.trading_days[0:time + 1]
        x = []
        for trials in range(trials):
            prices=[np.random.choice(range(*price_range))]
            for day in range(time):
                prices.append(
                    prices[day] * np.exp(
                    (rate - .5*vol**2) * (1.0 / 365) +
                    vol * (1.0 / 365)**.5 * np.random.normal(0,1)))
            x.append(prices)
        data = pd.DataFrame(x).T
        data.index = dates
        super(MonteCarloSource, self).__init__(data, **kwargs)
```

```python
N = 100
mus = [10**(5.0* t/N - 1.0) for t in range(N)]

    

def optimal_portfolio(prices):
    n = len(prices.columns)
    ## CALCULATE RETURNS
    R = np.log(prices / prices.shift(1)).dropna()
    
    ## CONVERT TO MATRIX
    S = opt.matrix(R.cov().values)
    pbar = opt.matrix(R.mean())
    
    ## CONDITION OPTIMIZER
    G = -opt.matrix(np.eye(n))   # negative n x n identity matrix
    h = opt.matrix(0.0, (n ,1))
    A = opt.matrix(1.0, (1, n))
    b = opt.matrix(1.0)
    
    ## CALCULAT EFFICIENT FRONTIER WEIGHTS
    portfolios = [solvers.qp(mu*S, -pbar, G, h, A, b)['x'] 
                  for mu in mus]
    ## CALCULATE RISKS AND RETURNS FOR FRONTIER
    returns = [blas.dot(pbar, x) for x in portfolios]
    risks = [np.sqrt(blas.dot(x, S*x)) for x in portfolios]
    ## CALCULATE THE 2ND DEGREE POLYNOMIAL OF THE FRONTIER CURVE
    m1 = np.polyfit(returns, risks, 2)
    x1 = sqrt(m1[2] / m1[0])
    wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']
    return pd.Series(wt, index=prices.columns)


data = MonteCarloSource(trials=4)

prices = data.data
w = optimal_portfolio(prices)
print w
plot(prices.values)
```
