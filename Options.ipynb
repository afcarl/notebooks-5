```python
%pylab inline
import datetime
import pandas as pd
from pandas.io.data import Options
from zipline.utils.factory import load_bars_from_yahoo
```

```python
# 3000 business day window
lookback = 3000 * pd.tseries.offsets.BDay()

end = pd.Timestamp.utcnow()
start = end - lookback

data = load_bars_from_yahoo(stocks=['XLP', 'XLE'],
                            start=start, end=end)

# Stock Price Data
prices = data.minor_xs('price')
norm_prices = prices / prices.iloc[0]

norm_prices.plot()
```

```python
# Build  Regression models
y = prices.XLP
X = prices.XLE

models = {n: pd.ols(y=y, x=X, 
                    window=n, 
                    window_type='rolling', 
                    intercept=False)
          for n in xrange(10, 161, 50)
}

betas = pd.DataFrame({n: models[n].beta['x'] for n in models})

fitted_X = betas * X

spreads = y - fitted_X

spreads.plot(figsize=(13,9))
```

```python


import math
from scipy.stats import norm
  
class Option:
    def __init__(self, s = 100, k = 100, t = 1, rf = 0.01, vol = 0.3):
        self.k = float(k)
        self.s = float(s)
        self.rf = float(rf)
        self.vol = float(vol)
        self.t = float(t)
  
    def get_call(self):
        d1 = ( math.log(self.s/self.k) + ( self.rf + math.pow( self.vol, 2)/2 ) * self.t ) / ( self.vol * math.sqrt(self.t) )
        d2 = d1 - self.vol * math.sqrt(self.t)
        self.call = ( norm.cdf(d1) * self.s - norm.cdf(d2) * self.k * math.exp( -self.rf * self.t ) )
        #put =  ( -norm.cdf(-d1) * self.s + norm.cdf(-d2) * self.k * math.exp( -self.rf * self.t ) ) 
        self.call_delta = norm.cdf(d1)
          
 
    def get_put(self):
        d1 = ( math.log(self.s/self.k) + ( self.rf + math.pow( self.vol, 2)/2 ) * self.t ) / ( self.vol * math.sqrt(self.t) )
        d2 = d1 - self.vol * math.sqrt(self.t)
        #call = ( norm.cdf(d1) * self.s - norm.cdf(d2) * self.k * math.exp( -self.rf * self.t ) )
        self.put =  ( -norm.cdf(-d1) * self.s + norm.cdf(-d2) * self.k * math.exp( -self.rf * self.t ) )
        self.put_delta = -norm.cdf(-d1) 
  
  
    def get_theta(self, dt = 0.01):
        self.t += dt
        self.get_call()
        self.get_put()
        after_call_price = self.call
        after_put_price = self.put
        self.t -= dt
        self.get_call()
        self.get_put()
        org_call_price = self.call
        org_put_price = self.put
 
        self.put_theta = (after_put_price - org_put_price) / dt * (-1)
        self.call_theta = (after_call_price - org_call_price) / dt * (-1)
 
    def get_all(self):
        self.get_call()
        self.get_put()
        self.get_theta()
        return self.put, self.call, self.put_delta, self.call_delta, self.put_theta, self.call_theta
 
```
